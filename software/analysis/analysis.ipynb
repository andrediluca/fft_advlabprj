```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def orderdered_output(x, npoints = 1024):
    x_new = [el.rstrip() for el in x] #remove newline
    x_int = [int(el, 16) for el in x_new[:-1]] #conversion to integer
    index = list(range(0,npoints)) #genersate index list
    index_bin = [f'{el:010b}' for el in index] #conversion of indeces to binary
    reversed_bin = [el[::-1] for el in index_bin] # 0 to 1 and 1 to 0 operation
    index_ordered = [int(el,2) for el in reversed_bin] # conversione to int
    ordered_y = [x_int[i] for i in index_ordered ] #reordering based on index_ordered
    return index, ordered_y
```

```python
f = open('output_re.mem', 'r')
output_fft = f.readlines()
f.close()
```

```python
x_, y = orderdered_output(output_fft)
```

```python
fig= plt.figure(figsize=(12,12))
plt.plot(x,y, 'bo')
```

# FFT with numpy

```python
import matplotlib.pyplot as plt
t = np.arange(256)
sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(t.shape[-1])
plt.plot(freq, sp.real, freq, sp.imag)
plt.show()
```
